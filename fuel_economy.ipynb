```python
import pandas as pd
import os
```

```python
fuel_econ_df = pd.read_csv('OGResources/fuel_economy.csv')
```

```python
fuel_econ_df.head()
```

```python
clean_fuel_econ_df = fuel_econ_df[["year",
                                   "make",
                                   "model",
                                   "city08",
                                   "highway08",
                                   "comb08",
                                   "fuelType1",
                                   "fuelCost08",
                                   "VClass",
                                   "cylinders",
                                   "trany",
                                   "drive"]]
```

```python
clean_fuel_econ_df.head()
```

```python
print(len(clean_fuel_econ_df))
```

```python
clean_fuel_econ_rename = clean_fuel_econ_df.rename(columns={"city08":"city_mpg",
                                                            "highway08":"highway_mpg",
                                                            "comb08":"combined_mpg",
                                                            "fuelType1":"fuel_type",
                                                            "fuelCost08":"annual_fuel_cost",
                                                            "VClass":"vehicle_size_class",
                                                            "cylinders":"engine_cylinders",
                                                            "trany":"transmission",
                                                            "drive":"drive_axle_type"})
```

```python
clean_fuel_econ_rename.head()
```

```python
clean_fuel_econ_rename['model']=clean_fuel_econ_rename['model'].str.replace('4WD','').str.replace('2WD','')
clean_fuel_econ_rename
```

```python
clean_fuel_econ_rename.to_csv('CleanResources/clean_fuel_economy.csv', 
                  encoding='utf-8', index=False, header=True)
```
