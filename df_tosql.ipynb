```python
df2=pd.read_csv('CleanResources/cleaned_customer_data.csv')
df2.to_sql('customer_data', engine, index=False)
```

```python
df3=pd.read_csv('CleanResources/clean_fuel_economy.csv')
df3.to_sql('fuel_economy_data', engine, index=False)

```
